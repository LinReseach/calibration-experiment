```python
import os
import re
import pandas as pd
filename='/Users/chenglinlin/ownCloud/BETA_AI_Gaze_Estimation_in_Human-Robot_Interaction (Projectfolder)/analized_data/analysis_dataset_neweth.csv'
df=pd.read_csv(filename)
df=df[(df['corrResp']==True)&(df['error_eth']<30)&(df['error_l2cs']<30)] 
df=df[df['person_ID'].isin(['2','4','12','13','14','15','18','19','20','21'])]
df
```

```python
df['3D_gaze']=df['3D_gaze'].str.split(',')
df['yp_gaze']=df['yp_gaze'].str.split(',')
df['yp_l2cs']=df['yp_l2cs'].str.split(',')
df['yp_eth']=df['yp_eth'].str.split(',')
def get_yaw_pitch(row):
    row['x']=round(float(row['3D_gaze'][0]),3)
    row['y']=round(float(row['3D_gaze'][1]),3)
    row['z']=round(float(row['3D_gaze'][2]),3)
    row['yaw']=round(float(row['yp_gaze'][0]),3)/3.14*180
    row['pitch']=round(float(row['yp_gaze'][1]),3)/3.14*180
    row['yaw_l2cs']=round(float(row['yp_l2cs'][0]),3)/3.14*180
    row['pitch_l2cs']=round(float(row['yp_l2cs'][1]),3)/3.14*180
    row['yaw_eth']=round(float(row['yp_eth'][0]),3)/3.14*180
    row['pitch_eth']=round(float(row['yp_eth'][1]),3)/3.14*180
    return row    

df=df.apply(get_yaw_pitch,axis='columns')
```

```python
df['yaw_error_l2cs']=df['yaw']-df['yaw_l2cs']
df['pitch_error_l2cs']=df['pitch']-df['pitch_l2cs']
df['yaw_error_eth']=df['yaw']-df['yaw_eth']
df['pitch_error_eth']=df['pitch']-df['pitch_eth']
p=[2,4,12,13,14,15,18,19,20,21]
i=0
ay=df[(df['4k']==False)&(~df['person_ID'].isin([p[i],p[i+1]]))]['yaw_error_eth'].mean()
ap=df[(df['4k']==False)&(~df['person_ID'].isin([p[i],p[i+1]]))]['pitch_error_eth'].mean()
(ay,ap),(ay,ap)
df
```

```python
import numpy as np
def sphere2cartesian(row):
    x0=row['yaw']
    x1=row['pitch']
    row['x']=np.cos(x1) * np.sin(x0)
    row['y']=np.sin(x1)  
    row['z']=-np.cos(x0) * np.cos(x1)
    return row

def error3d(a,b):
    a=a.to_numpy()
    b=b.to_numpy()
    e=[]
    for i in range(len(a)):
        error = np.arccos(np.dot(a[i], b[i]))
        e.append(error)
    dfe = pd.DataFrame(e)*180/3.1415926
    return dfe

def comp(idealgaze,pregaze,offset_y,offset_p):  
    pregaze['pitch']=pregaze['pitch']/180*3.14+offset_p/180*3.14#just test 0.122(7 degree) 0,105(6 degree)
    pregaze['yaw']=pregaze['yaw']/180*3.14+offset_y/180*3.14#just test 0.122(7 degree) 0,105(6 degree)

    pregaze=pregaze.apply(sphere2cartesian,axis='columns')  

    error=error3d(pregaze[['x','y','z']],idealgaze[['x','y','z']])
    return error.mean()
```
